```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Function to display our image and convert it to RBG image, give it an aspect ratio and a title

```python
def imshow(image=None,title = "Image",size = 10):
    w,h = image.shape[0],image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize = ((size * aspect_ratio),size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

### CV imread to read images

```python
img = cv2.imread("Lady.png")
imshow(img, title = "Temiloluwa")
```

### cv reading images without imshow function

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

### cv convert color to grayscale image

```python
gray_scale = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
imshow(gray_scale)
```

```python
# second method of viewing imags in grayscale
img2 = cv2.imread("Lady.png",0)
imshow(img2, title = "Grayscale")
```

### Image shapes Normal images have 3 dimension shapes(Red,Green,Blue) convolution

```python
print("Normal image shape is: ", img.shape)
print("Gray scale image shape is: ", gray_scale.shape)
```

### Splitting and merging images into BGR images as cv2 original form of viewing images is in BGR

```python
B,G,R = cv2.split(img)
imshow(R,"RED")
imshow(B,"BLUE")
imshow(G,"GREEN")
```

### Merging takes the array of all the color channel(BGR)

```python
zeros = np.zeros(img.shape[:2],dtype = "uint8")
imshow(cv2.merge([zeros,zeros,R]),"RED",5)
imshow(cv2.merge([zeros,G,zeros]),"GREEN",5)
imshow(cv2.merge([B,zeros,zeros]),"BLUE",5)
```

```python
# Splitted images in color channel marged back into one
merged = cv2.merge([B,G,R])
imshow(merged,title= "Merged Image")
```

### HSV(HUE,SATURATION,VALUE)
HUE determies color differences<br>
SATURATION determines how deep or bright the color is<br>
VALUE determines the intensity of the image

```python
hsv_image = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
imshow(hsv_image,"HSV IMAGE",5)
```

```python
im3 = cv2.cvtColor(hsv_image,cv2.COLOR_HSV2BGR)
imshow(im3)
```

### Viewing images with matplotlib plt.imshow giving us a BGR image compared to the function imshow image giving us the original image

```python
plt.imshow(im3)
imshow(im3,size=5)
```

```python
imshow(hsv_image[:,:,0],"Hue",5)
imshow(hsv_image[:,:,1],"Saturation",5)
imshow(hsv_image[:,:,2],"Value",5)
```

#### Creating line on our image
<em>cv2.line(image, starting cordinates, ending cordinates,color,thickness)

```python
cv2.line(image,(0,0),(511,511),(255,127,0),5)
imshow(image)
```

#### Creating rectangle on our image
<em>cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)

```python
image = np.zeros((512,512,3), np.uint8)

cv2.rectangle(image,(100,100),(300,250),(127,50,127),10)
imshow(image,"Black Canvas With Pink Rectangle",5)
```

```python
cv2.line(gray_image,(0,0),(511,511),(255,127,0),5)
imshow(gray_image,size=5)
```

#### Creating circle on our image
<em>cv2.circle(image,center,radius,color,fill)

```python
image = np.zeros((512,512,3),np.uint8)
cv2.circle(image,(350,350),100,(15,150,50),-1)
imshow(image,size=5)
```

#### Creating polylines on our image
<em>cv2.polylines(image,points,Closed?,color,thickness)

```python
pts = np.array([[10,50], [400,50], [90,200], [50,500]],np.int32)

pts.shape
```

```python
pts = pts.reshape((-1,1,2))
pts.shape
```

```python
image = np.zeros((512,512,3),np.uint8)

cv2.polylines(image, [pts], True, (220,225,255),3)
imshow(image,size=5)
```

#### Rotating images
<em>cv2.getRotationMatrix2D(rotation_center_x,rotation_center_y,angle of rotation,scale)

```python
height, width = image.shape[:2]

rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2),90,1)

rotated_image = cv2.warpAffine(image,rotation_matrix,(width,height))
imshow(rotated_image)
```

```python
rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2),90,0.5)

rotated_image = cv2.warpAffine(image,rotation_matrix,(width,height))
imshow(rotated_image)
```

```python
rotated_image = cv2.transpose(image)

imshow(rotated_image,"ROTATED WITH TRANSPOSE")
```

```python
flipped = cv2.flip(image,1)
imshow(flipped,"HORIZONTAL FLIP")
```
